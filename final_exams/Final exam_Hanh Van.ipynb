```python
#Question 1.b
df['Practice Type'].unique()
```

```python
def clean_practice_type(text):
    text = str(text)
    if text[0:3] == '17J':
        return '17J'
    if text[0:3] == '17C':
        return '17C'
    if text[0:2] == '2C':
        return '2C'
    return()

df['Practice Type'] = df['Practice Type'].apply(clean_practice_type)
df.head()
```

```python
df['Practice Type'].unique()
```

```python
#Question 1.c
df['Postcode'].unique()
```

```python
def clean_postcode(text):
    text = str(text)
    if ' ' in text:
        return text.replace(' ','-')
    if '-' in text:
        return text
    else:
        return text[:-3] + '-' + text[-3:]
    return text[1]

df['Postcode'] = df['Postcode'].apply(clean_postcode)
df.head()
```

```python
df['Postcode'].unique()
```

```python
#Question 1.d
df['Public'].unique()
```

```python
def clean_public(text):
    text = str(text)
    if text[0] == 'Y':
        return 'Y'
    return 'N'

df['Public'] = df['Public'].apply(clean_public)
df.head()
```

```python
df['Public'].unique()
```

```python
#Question 1.e
df = df.rename(columns={'Practice Code': 'PCode', 'Practice List Size': 'PSize'})
df.head()
```

```python
#Question 1.f
mean_psize = int(df['PSize'].mean())

def clean_psize(size):
    size = int(size)
    if size < 0:
        return mean_psize
    return size
    
df['PSize'] = df['PSize'].apply(clean_psize)
df.head()
```

```python
#Question 1.g
def capname(text):
    text = str(text)
    return text.title()
        
df['Practice Name'] = df['Practice Name'].apply(capname)
df.head()
```

```python
#Question 1.h
df['Month'] = df['Month_Year'].str.split('_').str.get(0)
df['Year'] = df['Month_Year'].str.split('_').str.get(1)
df.head()
```

```python
df.to_csv('cleaned.csv')
```
